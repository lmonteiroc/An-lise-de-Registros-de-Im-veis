## Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
dados['valor bruto'] = dados['Valor'] + dados['IPTU'] + dados['Condominio']
```

```python
dados.head(10)
```

```python
dados['valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
dados['valor m2'] = dados['Valor'] / dados['Area'].round(2)
```

```python
dados.head(10)
```

```python
dados['Valor bruto m2'] = dados['Valor']/dados['Area']
```

```python
dados['Valor bruto m2'] = dados['Valor bruto m2'].round(2)
```

```python
dados.head(10)
```
